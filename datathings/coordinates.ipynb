```python
ind = pd.read_excel('Important_child_qualities_unselfishness.xlsx')
print(ind.shape)
ind.head()
```

```python
countries = []
avoid = ['Value', 'TOTAL']
for col in ind:
    if col not in avoid:
        countries.append(col)
```

```python
print(countries)
```

```python
from bs4 import BeautifulSoup  
import requests
```

```python
baseurl = 'https://www.geodatos.net/en/coordinates/'
coord_map = {}

for country in countries:
    url = baseurl + country
    try:
        r = requests.get(url)
        soup = BeautifulSoup(r.content, 'html5lib')
        ps = soup.select('p', attr = {"class":'text-info nomargin'})
        text = ps[3].text.split(',')
        x = text[0]
        y = text[1]
        coord_map[country] = (x,y)
    except:
        print("Access for country: {} did not work".format(country))
```

```python
coord_map
```

```python
country_df = pd.DataFrame()

for country in coord_map:
    lat, long = coord_map[country]
    country_df = country_df.append({"Country":country, "Latitude": lat, "Longitude": long}, ignore_index = True)
    
```
