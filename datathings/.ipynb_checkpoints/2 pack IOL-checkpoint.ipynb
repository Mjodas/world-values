```python
import os
import pandas as pd
```

```python
path = './wave5IIL/'
icq_file = './wave5ICQ/wave5-clean.xls'
```

```python
new_columns = ['IIL: Family', 'IIL: Work','IIL: Religion','IIL: Politics',]
```

```python
for file in os.listdir(path):
    print(file)
```

```python
result = pd.DataFrame()
icq = pd.read_excel(icq_file)
```

```python
for file, col in zip(os.listdir(path), new_columns):
    
     if file == '.DS_Store':
        continue
    
    print(file)
    current_df = pd.read_excel(path+file)
    
    vi = current_df.iloc[0][1:]
    i = current_df.iloc[1][1:]
    data = i + vi

    result[col] = data
    print(result.shape)
```

```python
icq = icq.append(result)
```
